## データの場所、ラベルの追加

```python
df2['label'].value_counts()
```

```python
df2['label'].unique()
```

```python
def plus(df):
    if df['label'] == 'COVID':
        name = '/content/COVID-19_Radiography_Dataset/COVID-19_Radiography_Dataset/COVID' + '/' + df['file_name']
    elif df['label'] == 'Lung_Opacity':
        name = '/content/COVID-19_Radiography_Dataset/COVID-19_Radiography_Dataset/Lung_Opacity' + '/' + df['file_name']
    elif df['label'] == 'Normal':
        name = '/content/COVID-19_Radiography_Dataset/COVID-19_Radiography_Dataset/Normal' + '/' + df['file_name']
    else:
        name = '/content/COVID-19_Radiography_Dataset/COVID-19_Radiography_Dataset/Viral Pneumonia' + '/' + df['file_name']
    return name
```

```python
df2['path'] = df2.apply(plus, axis=1)
```

```python
df2['path'].unique()
```

```python
def label_num(df):
    if df['label'] == 'COVID':
        res = 0
    elif df['label'] == 'Lung_Opacity':
        res = 1
    elif df['label'] == 'Normal':
        res = 2
    else:
        res = 3
    return res
```

```python
df2['label_num'] = df2.apply(label_num, axis=1)
```
